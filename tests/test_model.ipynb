# Test Model Script
This notebook tests the YOLO model specified by the path. We have optimized it to load faster by disabling telemetry and internet checks.

```python
import time
import torch
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
from ultralytics import settings

settings.update({'sync': False})

%matplotlib inline
```

```python
model_path = r'C:\@dio\01_BINUS\04_Semester_4\03_Computer_Vision\02_Work\AOL\cv-code\models\v11-optimized-fast\weights\best.pt'

print("Initializing PyTorch...")
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

print(f"\nLoading model from: {model_path}")
start_time = time.time()

model = YOLO(model_path)
model.to(device)

end_time = time.time()
print(f"Model loaded successfully in {end_time - start_time:.2f} seconds!")
```

```python
image_path = 'data/test_image.jpg'

try:
    results = model(image_path)

    res_plotted = results[0].plot()

    plt.figure(figsize=(12, 12))
    plt.imshow(cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title('Inference Results')
    plt.show()
except Exception as e:
    print(f"Error running inference: {e}\nPlease ensure the image_path is correct and points to a valid image.")
```
